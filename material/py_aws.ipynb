```python
'''
用 request 隨便爬個網頁，
並把內存在當前目錄，

檔案名子叫做 lbh.html

python request to get a html example

'''
# 引入 requests 模組
import requests

# 使用 GET 方式下載普通網頁
r = requests.get('https://www.cxcxc.io/')

# 引出文字
print(r.text)

# 開檔案寫入
f = open("lbh.html", "a", encoding="utf8")
f.write(r.text)
f.close()

```

```python
'''
用boto3 上傳檔案到s3

boto3 upload object to s3 example

'''
# 跟任何服務溝通之前，都得有個客戶端
# 跟s3服務溝通，也要有個客戶端
s3_client = boto3.client('s3')

# upload_file 方法
# 可以上傳檔案
# 第一個參數是本地檔案的位置
# 第二個參數是籃子的位置
# 第三個參數是物件的位置

response = s3_client.upload_file("lbh.html", "iii-tutorial-v2", "student12/lbh.html")
print(response)
```

```python
'''
用boto3 瀏覽特定桶子內的物件 lbh.html

boto3 list object from specific bucket
'''

response = s3_client.list_objects_v2(
    Bucket='iii-tutorial-v2',
    MaxKeys=123,
    Prefix='student999/',
)

print(response)
```

```python
'''
用boto3 下載特定桶子內的物件 lbh.html

boto3 download object from bucket example

'''

import boto3

# 生成一個客戶端
s3 = boto3.client('s3')

# 第一個參數是桶子名
# 第二個參數是物件名
# 第三個參數是存回本地端之後，叫什麼名子
# s3.download_file('BUCKET_NAME', 'OBJECT_NAME', 'FILE_NAME')
s3.download_file('iii-tutorial-v2', 'student12/lbh.html', 'harveyWu.html')
```

```python
'''
用boto3 刪除特定桶子內的物件 lbh.html

boto3 delete object from bucket example

'''

# 生成一個客戶端
s3 = boto3.client('s3')

response = s3.delete_object(
    Bucket='iii-tutorial-v2',
    Key='student12/lbh.html',
)

print(response)

```
